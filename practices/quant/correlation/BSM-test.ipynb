```python
def get_g(a,b,c):
    return sp.sqrt(b**2 - 4.0*a*c)
    
def get_Riccati_solution(a,b,c,tau,f0):
    g = get_g(a,b,c)
    ex = sp.exp(g*tau) - 1
    num = 2*g*f0 + ((b+g)*f0+2*c)*ex
    denom = 2*g - (2*a*f0+b-g)*ex
    return num/denom

def get_Riccati_solution_vec(a,b,c,tau,f0):
    n = len(a)
    res = np.zeros(n)
    for i in range(n):
        res[i] = get_Riccati_solution(a[i],b[i],c[i],tau,f0[i])
    return res

def get_int_Riccati_solution(a,b,c,tau,riccati_f0):
    g = get_g(a,b,c)
    ex = sp.exp(g*tau) - 1
    term1 = (-b+g)*tau/(2.0*a)
    term2 = np.log(2*g/np.abs(2*g+(g-b-2*a*riccati_f0)*ex))/a
    return term1 + term2

def get_int_Riccati_solution_vec(a,b,c,tau,riccati_f0):
    n = len(a)
    res = np.zeros(n)
    for i in range(n):
        res[i] = get_int_Riccati_solution(a[i],b[i],c[i],tau,riccati_f0[i])
    return res

def get_E_v(t,alp,beta,sigma,v0):
    res = np.exp(-beta*t)*(v0 - alp/beta) + alp/beta
    return res    

def get_E_vAvB(t,alp,beta,sigma,v0):
    res = np.exp(-beta*t)*(v0 - alp/beta) + alp/beta
    return res.prod()

def get_int_E_vAvB(t,alpha,beta,sigma,v0):
    """
    Vectorized implementation
    """
    res = -(v0-alpha/beta).prod()*(np.exp(-beta.prod()*t)-1)/beta.sum() - np.sum((v0-alpha/beta)*np.flip(alpha)/beta.prod()*(np.exp(-beta*t)-1)) + alpha.prod()/beta.prod()*t
    return res

def get_CIR_Sample(NoOfPaths,kappa,gamma,vbar,s,t,v_s):
    delta = 4.0 *kappa*vbar/gamma/gamma
    c= 1.0/(4.0*kappa)*gamma*gamma*(1.0-np.exp(-kappa*(t-s)))
    kappaBar = 4.0*kappa*v_s*np.exp(-kappa*(t-s))/(gamma*gamma*(1.0-np.exp(-kappa*(t-s))))
    sample = c* np.random.noncentral_chisquare(delta,kappaBar,NoOfPaths)
    return sample
```

```python
corr = 0.5
params = {
# chf params    
    'tau': 1, # investment time
    'h': 30, # horizon in the future
    'u': np.array([1.0,1.0]), # weights for price
    'si': np.array([0.0181427,0.0422960,0.034]), # adjusted si[2]
    'sigma': np.array([0.232,0.099]),
    'x0': np.array([0.05095958,0.06725220,0.00961570]),
    's0': np.zeros(2),
    'correlation': np.matrix([[1, corr],
                             [corr, 1]])
}
params['r'] = 0.004965438104596101*np.ones(len(params['s0']))
params['L'] = np.linalg.cholesky(params['correlation'])
```

# Monte Carlo

```python
# input parameters
p = params
tau = p["tau"]
h = p["h"]
T = h + tau
sigma = p["sigma"]
s0 = p["s0"]
L = p["L"]
r = p['r']

# simulation parameters
np.random.seed(1)
num_steps = int(T*500)
num_paths = 15000
res = np.zeros((num_steps, num_paths))
dt = T/num_steps
t_sim = np.linspace(0.0,T,num_steps+1)
# Initialize states

s = np.zeros([num_steps+1, len(s0), num_paths]) # log price

# Initialization
for j in range(len(s0)):
    s[0,j,:] = s0[j]
    
for i in range(num_steps):
    # noncentral chisquare processes
    # Volatility
    dW_indep = np.random.normal(0.0,1.0,(2,num_paths))
    dW_indep = (dW_indep-dW_indep.mean(axis=1).reshape(-1,1))/dW_indep.std(axis=1).reshape(-1,1) # normalize
    dW = L*dW_indep*np.sqrt(dt) # each row (sA sB vA vB) 
    # Stock price
    for j in range(len(s0)): # j in (A,B)
        drift_term = (r[j] - 0.5*np.power(sigma[j],2))*dt
        dWs = dW[j,:] # A or B
        vol_term = sigma[j]* dWs
        s[i+1,j,:] = s[i,j,:] + drift_term + vol_term
    
```

# Covariance from MC

```python
h_steps = int(h/T*num_steps)
tau_steps = int(tau/T*num_steps)
mc_covs = np.zeros(h_steps)
for h_step in range(h_steps):
    h_tau_step = h_step+tau_steps
    mc_du = (s[h_tau_step,:,:]-s[h_step,:,:]).mean(axis=1)
    mc_duAuB = (s[h_tau_step,:,:]-s[h_step,:,:]).prod(axis=0).mean()
    mc_covs[h_step] = mc_duAuB - mc_du.prod()
    
x_axis = np.array(range(h_steps))/h_steps*h
plt.plot(x_axis,mc_covs)
plt.title('Covariance in Simulation')
```

# Characteristic Function for Correlated GBM

```python
def get_chf(p,tau=None,u=None):
    if tau is None:
        tau = p['tau']
    if u is None:
        u = p['u']
    corr = p['correlation'][0,1]
    B = u
    dAdtau = 0.5*np.power(p['sigma']*u,2).sum() + (p['sigma']*u).prod()*corr + np.dot(u,p['r']-0.5*np.power(p['sigma'],2))
    A = tau*dAdtau
    ret = np.exp(A + np.dot(B,p['s0']))
#     print(tau,dAdtau,B)
    return ret, A, B
```

```python
def get_return_chf(p,tau=None,h=None,u=None):
    if tau is None:
        tau = p['tau']
    if h is None:
        h = p['h']
    if u is None:
        u = p['u']
#     T = tau + h
    # ChF for return
    chf,A,B = get_chf(p,tau,u)
    chf2,A2,B2 = get_chf(p,h,B)
    return np.exp(A)*chf2,A2,B2,chf,A,B
```

```python
def get_return_correlation(p,tau=None,h=None):
# return correlation
    if tau is None:
        tau = p['tau']
    if h is None:
        h = p['h']
    u = np.zeros(len(p['s0']))
    corr = p['correlation'][0,1]
    dAt = 0.5*np.power(p['sigma']*u,2).sum() + (p['sigma']*u).prod()*corr + np.dot(u,p['r']-0.5*np.power(p['sigma'],2))
    Psi = np.exp(dAt*tau)
    dAtdu1 = u[0]*p['sigma'][0]**2 + p['sigma'].prod()*u[1]*corr + r[0] - 0.5*p['sigma'][0]**2
    dAtdu2 = u[1]*p['sigma'][1]**2 + p['sigma'].prod()*u[0]*corr + r[1] - 0.5*p['sigma'][1]**2
    d2Atdu1du2 = p['sigma'].prod()*corr

    dPsidu1 = (tau*dAtdu1)*Psi
    dPsidu2 = (tau*dAtdu2)*Psi
    d2Psidu1du2 = tau*(d2Atdu1du2 + dAtdu1*dAtdu2)*Psi
    ret_corr = d2Psidu1du2 - dPsidu1*dPsidu2
    print(d2Psidu1du2,dPsidu1,dPsidu2)
    return ret_corr
```
